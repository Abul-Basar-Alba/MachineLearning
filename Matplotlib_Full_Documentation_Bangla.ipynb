## üìä 4. Plot ‡¶ó‡ßÅ‡¶≤‡ßã‡¶∞ ‡¶™‡ßÇ‡¶∞‡ßç‡¶£ ‡¶§‡¶æ‡¶≤‡¶ø‡¶ï‡¶æ ‡¶ì ‡¶ï‡ßã‡¶° ‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£

### ‚úÖ A. Line Plot
```python
plt.plot([1,2,3], [4,5,6], color='blue', marker='o', linestyle='--')
```

### ‚úÖ B. Bar Chart
```python
plt.bar(['A','B','C'], [10,20,15], color='orange')
```

### ‚úÖ C. Horizontal Bar Chart
```python
plt.barh(['A','B','C'], [10,20,15])
```

### ‚úÖ D. Scatter Plot
```python
plt.scatter([1,2,3], [4,6,5], color='green', marker='x')
```

### ‚úÖ E. Histogram
```python
data = np.random.randn(1000)
plt.hist(data, bins=50, color='gray')
```

### ‚úÖ F. Pie Chart
```python
plt.pie([20, 30, 50], labels=["Python", "Java", "C++"], autopct='%1.1f%%')
```

### ‚úÖ G. Stack Plot
```python
days = [1, 2, 3, 4]
eating = [1, 2, 3, 4]
sleeping = [2, 3, 4, 3]
plt.stackplot(days, eating, sleeping, labels=['Eating','Sleeping'])
plt.legend()
```

### ‚úÖ H. Box Plot
```python
data = np.random.normal(0, 1, 100)
plt.boxplot(data)
```

### ‚úÖ I. Error Bar Plot
```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.errorbar(x, y, yerr=0.1)
```

### ‚úÖ J. Fill Between
```python
x = np.linspace(0, 1, 100)
y1 = np.sin(2*np.pi*x)
plt.fill_between(x, y1, color='skyblue')
```

## üîÑ 6. Subplots ‡¶è‡¶¨‡¶Ç Multi-plot

### ‚úÖ A. Basic Subplot
```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].bar(['A', 'B'], [3, 5])
```

### ‚úÖ B. Shared Axis
```python
fig, axs = plt.subplots(2, sharex=True)
```

## üéû 9. Animation (FuncAnimation ‡¶¶‡¶ø‡ßü‡ßá)
```python
from matplotlib.animation import FuncAnimation

def animate(i):
    y = np.random.randn(100)
    plt.cla()
    plt.plot(y)

ani = FuncAnimation(plt.gcf(), animate, interval=1000)
plt.show()
```

## üåê 10. 3D Plotting (mpl_toolkits)
```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1,2,3]; y = [4,5,6]; z = [7,8,9]
ax.scatter(x, y, z)
```
