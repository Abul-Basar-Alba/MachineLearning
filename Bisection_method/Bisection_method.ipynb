# Bisection Method Example

## Question: Determine the root of the given equation x2-3 = 0 for x ∈ [1, 2]

### Python Code (Graphical + Tabular Format)

```python
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 - 3

# Bisection Method Function
def bisection(a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Bisection method fails. f(a) and f(b) must have opposite signs.")
        return None

    print("Iter\t   a\t\t   b\t\t   t\t   f(t)")
    print("-" * 60)
    steps = []
    
    for i in range(1, max_iter + 1):
        t = (a + b) / 2.0
        ft = f(t)
        steps.append((i, a, b, t, f(a), f(b), ft))
        print(f"{i:2d}\t{a:.6f}\t{b:.6f}\t{t:.6f}\t{ft:.6f}")

        if abs(ft) < tol:
            break
        elif f(a) * ft < 0:
            b = t
        else:
            a = t

    return t, steps

# Call the function with interval [1, 2]
root, steps = bisection(1, 2)

# Tabular output already printed
# Now plot the function and the iterations
x_vals = [x/100.0 for x in range(100, 201)]
y_vals = [f(x) for x in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x² - 3')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Graphical Representation of Bisection Method")
plt.xlabel("x")
plt.ylabel("f(x)")

# Plot iterations
for i, a, b, t, fa, fb, ft in steps:
    plt.plot(t, f(t), 'ro')
    plt.text(t, f(t), f'{i}', fontsize=8)

plt.grid(True)
plt.legend()
plt.show()
```

# Example-1
## 1. Find a root of an equation f(x)=x3-x-1 using Bisection method

### Python Code (Graphical + Tabular Format)

```python
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**3 - x - 1

# Bisection Method Function
def bisection_method(a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Invalid interval. f(a) and f(b) must have opposite signs.")
        return None, []

    steps = []
    print(f"{'Iter':<5}{'a':<10}{'f(a)':<12}{'b':<10}{'f(b)':<12}{'c':<10}{'f(c)':<12}{'Update'}")
    print("-" * 80)

    for n in range(1, max_iter + 1):
        c = (a + b) / 2.0
        fa = f(a)
        fb = f(b)
        fc = f(c)

        update = ""
        if abs(fc) < tol:
            update = "Root found"
        elif fa * fc < 0:
            b = c
            update = "b = c"
        else:
            a = c
            update = "a = c"

        steps.append((n, a, fa, b, fb, c, fc, update))

        print(f"{n:<5}{a:<10.5f}{fa:<12.5f}{b:<10.5f}{fb:<12.5f}{c:<10.5f}{fc:<12.5f}{update}")

        if abs(fc) < tol:
            break

    return c, steps


# Run the example: x^3 - x - 1 = 0 on interval [1, 2]
root, iteration_data = bisection_method(1, 2)

# Plotting function and midpoints
x_vals = [x / 100.0 for x in range(90, 210)]
y_vals = [f(x) for x in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x³ - x - 1', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='green', linestyle='--', label=f'Root ≈ {root:.5f}')

# Plot all midpoint guesses
for i, a_i, fa_i, b_i, fb_i, c_i, fc_i, update in iteration_data:
    plt.plot(c_i, fc_i, 'ro')
    plt.text(c_i, fc_i, f'{i}', fontsize=8)

plt.title("Bisection Method: f(x) = x³ - x - 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Find the root of f(x) = x³ - 5 in the interval [1, 2] using the bisection method up to 3 iterations.

### Python Code (Graph + Table)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the function
def f(x):
    return x**3 - 5

# Bisection method
def bisection(a, b, iterations):
    table = []
    for i in range(1, iterations + 1):
        c = (a + b) / 2
        fa, fb, fc = f(a), f(b), f(c)
        update = ""
        if fa * fc < 0:
            b = c
            update = "b = c"
        else:
            a = c
            update = "a = c"
        table.append([i, round(a, 5), round(fa, 5), round(b, 5), round(fb, 5),
                      round(c, 5), round(fc, 5), update])
    return table

# Initial values and iterations
a, b = 1, 2
iterations = 3
data = bisection(a, b, iterations)

# Display table using pandas
df = pd.DataFrame(data, columns=["Iter", "a", "f(a)", "b", "f(b)", "c", "f(c)", "Update"])
print(df)

# Plot function and root interval
x = np.linspace(0, 2.5, 400)
y = f(x)
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x) = x^3 - 5', color='blue')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(1, color='green', linestyle='--', label='Initial a = 1')
plt.axvline(2, color='red', linestyle='--', label='Initial b = 2')
plt.scatter(df["c"], df["f(c)"], color='orange', label='Midpoints c')
plt.title('Bisection Method Graph (f(x) = x^3 - 5)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
