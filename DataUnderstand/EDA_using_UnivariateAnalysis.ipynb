```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
sns.countplot(x=df['Survived'])
plt.show()
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')  # color change
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
counts = df['Pclass'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Survival Distribution")
plt.show()
```

```python
plt.hist(df['Age'],bins=60)
```

```python
sns.histplot(data=df, x='Age', bins=30, kde=True, color='purple')
plt.title("Age Distribution (Histogram + KDE)")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True, palette='Set2')
plt.title("Age Distribution by Survival (Histogram + KDE)")
plt.show()

```

```python
sns.boxplot(x='Age', data=df)
plt.title("Age Distribution (Box Plot)")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.boxplot(x='Sex', y='Fare', hue='Survived', data=df, palette='Set3')
plt.title("Fare Distribution by Sex and Survival")
plt.show()
```
